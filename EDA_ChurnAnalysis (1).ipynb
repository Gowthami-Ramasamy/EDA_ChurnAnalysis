```python
df.tenure.value_counts().plot(kind='bar')
```

```python
bins = [0, 20, 40, 60, 80]
labels = ['0-20', '21-40', '41-60', '61-80']

# Create the 'tenure_bins' column using pd.cut
df['tenure_bins'] = pd.cut(df['tenure'], bins=bins, labels=labels, include_lowest=True)
```

```python
df.head(5)
```

```python
df.tenure_bins.value_counts().plot(kind='bar')
```

```python
df.drop(columns=['customerID', 'tenure'], inplace=True)
```

```python
df.head(5)
```

```python
for i,predictor in enumerate(df.drop(columns=['Churn','TotalCharges','MonthlyCharges'])):
  plt.figure(i, figsize=(6,4))

  ax = sns.countplot(data=df,x=predictor,hue='Churn')

  # Add count labels to the bars
  for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 5), textcoords = 'offset points')

  ax.set_xlabel('Predictor')
  ax.set_ylabel('Count')
  ax.set_title('Count Plot for ' + predictor + ' with Churn')
  plt.show()
```
